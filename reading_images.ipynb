### Now we will split and merge color channels

cv2.split() divides into single channel arrays

cv2.merge() merges several arrays to make multi channel

```python
#split the image into B,G,R componenets.
img_ferrari_bgr = cv2.imread('images/ferrari.jpg', cv2.IMREAD_COLOR)
b,g,r = cv2.split(img_ferrari_bgr)

#show the channels
plt.figure(figsize=[20,5])
plt.subplot(141); plt.imshow(r); plt.title('Red channel');
plt.subplot(142); plt.imshow(g); plt.title('Green channel');
plt.subplot(143); plt.imshow(b); plt.title('Blue channel');
#plt.Subplot displays a smaller plot 
#plt.imshow is the actual plot
#plt.title is the title

#merge channels
ferrari_merged = cv2.merge((b,g,r))
#show output
plt.subplot(144); plt.imshow(ferrari_merged[:,:,::-1]);plt.title('Merged output')
```

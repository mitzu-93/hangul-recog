### Now we will split and merge color channels

cv2.split() divides into single channel arrays

cv2.merge() merges several arrays to make multi channel

```python
#split the image into B,G,R componenets.
img_ferrari_bgr = cv2.imread('images/ferrari.jpg', cv2.IMREAD_COLOR)
b,g,r = cv2.split(img_ferrari_bgr)

#show the channels
plt.figure(figsize=[20,5])
plt.subplot(141); plt.imshow(r); plt.title('Red channel');
plt.subplot(142); plt.imshow(g); plt.title('Green channel');
plt.subplot(143); plt.imshow(b); plt.title('Blue channel');
#plt.Subplot displays a smaller plot 
#plt.imshow is the actual plot
#plt.title is the title

#merge channels
ferrari_merged = cv2.merge((b,g,r))
#show output
plt.subplot(144); plt.imshow(ferrari_merged[:,:,::-1]);plt.title('Merged output')
```

cv2.cvtColor() converts an image from one color space to another. 

#### Function Syntax:

dst = cv2.cvtColor(src, code)

dst: is the output image.

src: is the input image

code: is the color space conversion

```python
#this essentially reverses the channels to make it RGB and not BGR
ferrari_rgb = cv2.cvtColor(img_ferrari_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(ferrari_rgb)
```

```python
img_hsv = cv2.cvtColor(img_ferrari_bgr, cv2.COLOR_BGR2HSV)
#split img into bgr
h,s,v = cv2.split(img_hsv)

#show channels
plt.figure(figsize=[20,5])
#cmap gray actually makes it more clear what we are looking at.
plt.subplot(141);plt.imshow(h, cmap='gray'); plt.title('H channel');
plt.subplot(142);plt.imshow(s, cmap='gray'); plt.title('S channel');
plt.subplot(143);plt.imshow(v, cmap='gray'); plt.title('V channel');
plt.subplot(144);plt.imshow(ferrari_rgb, cmap='gray'); plt.title('Original');
```

```python
h_new = h+10
ferrari_merged = cv2.merge((h_new, s, v))
ferrari_rgb = cv2.cvtColor(ferrari_merged, cv2.COLOR_HSV2RGB)

#show the channels
plt.figure(figsize=[20,5])
plt.subplot(141); plt.imshow(h, cmap='gray'); plt.title('H channel');
plt.subplot(142); plt.imshow(s, cmap='gray'); plt.title('S channel');
plt.subplot(143); plt.imshow(v, cmap='gray'); plt.title('V channel');
plt.subplot(144); plt.imshow(ferrari_rgb, cmap='gray'); plt.title('Modified');
```

```python
# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (9.0, 9.0)
from IPython.display import Image
```

```python
image = cv2.imread('images/space_x.png', cv2.IMREAD_COLOR)
```

```python
plt.imshow(image[:,:,::-1])
```

```python
#draw a circle
imageCircle = image.copy()

cv2.circle(imageCircle, (780,480), 100, (0, 0, 255), thickness=5, lineType=cv2.LINE_AA);

#the image
plt.imshow(imageCircle[:,:,::-1])
```

```python
imageRectangle = image.copy()

cv2.rectangle(imageRectangle, (400,450), (600,100), (255, 0, 255), thickness=5, lineType=cv2.LINE_8)

plt.imshow(imageRectangle[:,:,::-1])
```

```python
imageText = image.copy()
text = 'Space X launch'
fontScale = 2.3
fontFace = cv2.FONT_HERSHEY_PLAIN
fontColor = (0, 255, 0)
fontThickness = 2

cv2.putText(imageText, text, (350, 550), fontFace, fontScale, fontColor, fontThickness, cv2.LINE_AA)

#display image
plt.imshow(imageText[:,:,::-1])
```

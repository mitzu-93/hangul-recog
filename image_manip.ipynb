```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
from IPython.display import Image
%matplotlib inline
```

```python
# read image as gray scale
ferrari = cv2.imread('images/ferrari.jpg', 0)

#set color map to gray scale
plt.imshow(ferrari, cmap='gray')
print(ferrari)
```

```python
# print shape so we know where we can edit single pixels
print(ferrari.shape)
```

```python
#rgb value of this particular pixel 
print(ferrari[300,300])
```

```python
#changing pixel values
ferrari_copy = ferrari.copy()
#this is now a splice of the pixel range.
ferrari_copy[300:310] = 200
ferrari_copy[310:320] = 255
ferrari_copy[320:330] = 0
ferrari_copy[330:340] = 100

#cannot really see since the image is quite large
plt.imshow(ferrari_copy, cmap='gray')
```

```python
ferrari_bgr = cv2.imread('images/ferrari.jpg', cv2.IMREAD_COLOR)
ferrari_bgr = ferrari_bgr[:,:,::-1]

plt.imshow(ferrari_bgr)
```

```python
cropped_region = ferrari_bgr[200:600, 200:600]
plt.imshow(cropped_region)
```

```python
resize_cropped_ferrari = cv2.resize(cropped_region, None, fx=2, fy=2)
plt.imshow(resize_cropped_ferrari)
```

```python
#using dsize
des_width = 100
aspect_ratio = des_width / cropped_region.shape[1]
des_height = int(cropped_region.shape[0] * aspect_ratio)
dim = (des_width, des_height)

#resize image
resize_cropped_ferrari = cv2.resize(cropped_region, dsize=dim, interpolation=cv2.INTER_AREA)
plt.imshow(resize_cropped_ferrari)
```

```python
des_height = 100
des_width = 200
dim = (des_height, des_width)

#resize background image to same size
resize_cropped_ferrari = cv2.resize(cropped_region, dsize=dim, interpolation=cv2.INTER_AREA)
plt.imshow(resize_cropped_ferrari)
```

```python
#swap channel order
resize_cropped_ferrari_2x = resize_cropped_ferrari[:,:,::-1]

#safve resized image to disk
cv2.imwrite('images/resize_cropped_ferrari_2x.png', resize_cropped_ferrari_2x)

#display the cropped and resized image
Image(filename='images/resize_cropped_ferrari_2x.png')
```

```python
hor_flip = cv2.flip(ferrari_bgr, 1)
vert_flip = cv2.flip(ferrari_bgr, 0)
both_flip = cv2.flip(ferrari_bgr, -1)

#show images
plt.figure(figsize=[18, 5])
plt.subplot(141); plt.imshow(hor_flip); plt.title('horizontal flip')
plt.subplot(142); plt.imshow(vert_flip); plt.title('vertical flip')
plt.subplot(143); plt.imshow(both_flip); plt.title('both flip')
plt.subplot(144); plt.imshow(ferrari_bgr); plt.title('original')
```
